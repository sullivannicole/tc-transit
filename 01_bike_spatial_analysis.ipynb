The above map gives a quick look at density of Nice Ride stations in Minneapolis. Let's get the quantitative numbers programmatically as well, using spatial joins. To do that, first, we'll convert the bike stations pandas dataframe to a geodataframe with a geometry column.

```python
stations_df = geopandas.GeoDataFrame(bike_stations[['name', 'empty_slots']], geometry = geopandas.points_from_xy(bike_stations.longitude, bike_stations.latitude, crs = 'epsg:4326'))
```

Now, let's perform a spatial join.

```python
mpls_comm.crs
```

```python
stations_df.crs
```

```python
stations_comms = geopandas.sjoin(stations_df,  mpls_comm, how = "inner", op = 'intersects')
```

## Which communities have the most bike stations?

```python
stations_comms.groupby('CommName') \
                          .count() \
                          .sort_values(by = 'empty_slots', ascending = False)
```

## Which communities have the most bikes?

```python
stations_comms.groupby('CommName') \
              .sum() \
              .sort_values(by = 'empty_slots', ascending = False)
```

We can see the orders are nearly identical, except for Camden and Southwest. While Southwest has fewer stations than Camden, the community actually has a few more bikes (n = 77) in total than Camden (n = 73).

## Which communities have the **highest bike density per square mile**?

```python
comm_crs = mpls_comm.copy()
comm_crs = comm_crs.to_crs({'init': 'epsg:3857'})
comm_crs['area_sqkm'] = comm_crs.area/10**6
```

```python
bike_comm_summed = stations_comms.groupby('CommName').sum()
```

```python
bike_density = comm_crs.join(bike_comm_summed['empty_slots'], on = 'CommName')
```

```python
bike_density['bikes_per_km'] = bike_density['empty_slots'] / bike_density['area_sqkm']
```

```python
bike_density.sort_values(by = 'bikes_per_km', ascending = False)
```

Fascinating! Looking only at quantity of bikes doesn't give us the full picture. Phillips falls 6th in terms of total bikes, but when we look at bikes per kilometer in each community, the Phillips neighborhood jumps up to 2nd-highest spot, with over 42 bikes per square kilometer!
