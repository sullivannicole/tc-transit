## Where are Nice Ride stations in relation to Minneapolis communities?

```python
bike_comms
```

The above map gives a quick look at density of Nice Ride stations in Minneapolis. Let's get the quantitative numbers programmatically as well, using spatial joins. To do that, first, we'll convert the bike stations pandas dataframe to a geodataframe with a geometry column.

```python
stations_df = geopandas.GeoDataFrame(bike_stations[['name', 'empty_slots']], geometry = geopandas.points_from_xy(bike_stations.longitude, bike_stations.latitude, crs = 'epsg:4326'))
```

Now, let's perform a spatial join.

```python
mpls_comm.crs
```

```python
stations_df.crs
```

```python
stations_comms = geopandas.sjoin(stations_df,  mpls_comm, how = "inner", op = 'intersects')
```

## Which communities have the most bike stations?

```python
stations_comms.groupby('CommName') \
                          .count() \
                          .sort_values(by = 'empty_slots', ascending = False)
```

## Which communities have the most bikes?

```python
stations_comms.groupby('CommName') \
              .sum() \
              .sort_values(by = 'empty_slots', ascending = False)
```

We can see the orders are nearly identical, except for Camden and Southwest. While Southwest has fewer stations than Camden, the community actually has a few more bikes (n = 77) in total than Camden (n = 73).

## Which communities have the **highest bike density per square mile**?

```python
comm_crs = mpls_comm.copy()
comm_crs = comm_crs.to_crs({'init': 'epsg:3857'})
comm_crs['area_sqkm'] = comm_crs.area/10**6

bike_comm_summed = stations_comms.groupby('CommName').sum()
bike_density = comm_crs.join(bike_comm_summed['empty_slots'], on = 'CommName')
bike_density['bikes_per_km'] = bike_density['empty_slots'] / bike_density['area_sqkm']
bike_density.sort_values(by = 'bikes_per_km', ascending = False)
```

Fascinating! Looking only at quantity of bikes doesn't give us the full picture. Phillips falls 6th in terms of total bikes, but when we look at bikes per kilometer in each community, the Phillips neighborhood jumps up to 2nd-highest spot, with over 42 bikes per square kilometer!

## The "last mile" problem: weekday transit and bike proximity

```python
trips_q = '''SELECT a.site_id, a.route, a.dir, a.ons, a.offs, b.geo_node_name, b.latitude, b.longitude, b.city, b.downtown
            FROM `tc-transit.transit.trips` a
            LEFT JOIN `tc-transit.transit.sites` b
            ON a.site_id = b.site_id
            
            WHERE a.schedule = "Weekday";'''

trips = client.query(trips_q).to_dataframe()
```

```python
trip_map = leafmap.Map(height = 900)
trip_map.add_data(data = trips, name = 'transit_trips')
trip_map.config = trip_config
```

```python
trip_map.config
```

```python
trips_mpls = trips.query('city == "Minneapolis"')
```

```python
trip_mpl_map = leafmap.Map(height = 1000)
trip_mpl_map.add_data(data = trips_mpls, name = 'mpls_transit_trips')
# trip_mpl_map.config = trip_mpl_config
```

```python
trip_mpl_map
```

```python
# Convert trips to geometry
trip_df = geopandas.GeoDataFrame(trips[['site_id', 'dir', 'geo_node_name','city', 'ons', 'offs']], geometry = geopandas.points_from_xy(trips.longitude, trips.latitude, crs = 'epsg:4326'))

# Get the nearest bike station to the transit stop
station_nearest = geopandas.sjoin_nearest(trip_df, stations_df, distance_col = 'distance')

# Filter down to Minneapolis locations only
mpls_nearest_bikes = station_nearest.query('city == "Minneapolis"')

# Reconnect the geometry for the bike stations to this df by the name of the station
mpls_near_stations = mpls_nearest_bikes.merge(stations_df[['name', 'geometry']], on = 'name', how = 'left')

# Transit geometry is now an object, convert to a geometry and store in GeoDataFrame
geox = geopandas.GeoSeries.from_wkt(mpls_near_stations['geometry_x'])
transit_pts = geopandas.GeoDataFrame({'station': mpls_near_stations['geo_node_name'], 'geometry': geox}, crs = 'EPSG:4326')
bike_pts = geopandas.GeoDataFrame({'geometry': mpls_near_stations['geometry_y']}, crs = 'EPSG:4326')

# Project into local coord system for more precise results
transit_pts = transit_pts.to_crs('EPSG:3594')
bike_pts = bike_pts.to_crs('EPSG:3594')
```

```python
transit_pts['distance_nearest_bike_station_m'] = transit_pts.distance(bike_pts)
```

```python
transit_pts
```

```python
near_bikes_map = leafmap.Map(height = 1000)
near_bikes_map.add_data(data = transit_pts, name = 'trips_nearest_bike_station')
# trip_map.config = trip_config2
```

```python
near_bikes_map
```

While it's obvious that transit stops downtown are the closest to bike stations, it's also clear there are other corridors along which bike stations have been put in quite close proximity to transit stops (Lake Street and Penn Ave are of particular note)
